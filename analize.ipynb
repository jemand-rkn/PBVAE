```python
import pandas as pd

df = pd.read_csv('./wandb_export_2026-02-01T15_41_43.499+09_00.csv')
df
```

```python
new_df = df.apply(lambda x: pd.Series(x.dropna().values))
new_df
```

```python
# new_dfã‹ã‚‰æœ«å°¾ãŒ'_MIN'ã¾ãŸã¯'_MAX'ã®ã‚«ãƒ©ãƒ ã‚’å‰Šé™¤
cols_to_drop = [col for col in new_df.columns if col.endswith('_MIN') or col.endswith('_MAX')]
filtered_df = new_df.drop(columns=cols_to_drop)
filtered_df
```

```python
# å„ã‚«ãƒ©ãƒ ã”ã¨ã«NaNã§ãªã„ãƒ‡ãƒ¼ã‚¿ã®æœ«å°¾ï¼ˆæœ€å¾Œã®å€¤ï¼‰ã‚’å–å¾—
last_valid_values = filtered_df.apply(lambda col: col.dropna().iloc[-1] if col.dropna().size > 0 else None)
last_valid_values
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ãƒ‡ãƒ¼ã‚¿ã®æº–å‚™ï¼ˆå¾Œã§å®Ÿéš›ã®å€¤ã‚’å…¥åŠ›ï¼‰
beta_values = np.array([0.001, 0.1, 1.0])  # Î²å€¤

# Test reconstruction lossï¼ˆå·¦ã®ãƒ—ãƒ­ãƒƒãƒˆç”¨ï¼‰
test_recon = {
    'Î²-vae': np.array([57.550373, 61.462469, 79.588030]),  # ä¾‹ï¼šå®Ÿéš›ã®å€¤ã‚’å…¥åŠ›
    'prior Î²-vae, dropout, kl att (classic)': np.array([73.718887, 78.124582, 94.689591]),
    'prior Î²-vae, dropout, kl att (quad)': np.array([72.952846, 76.395155, 92.400047]),
}

# Generalization gapï¼ˆå³ã®ãƒ—ãƒ­ãƒƒãƒˆç”¨ï¼‰
gen_gap = {
    'Î²-vae': np.array([1.582914, 1.575519, 2.663287]),
    'prior Î²-vae, dropout, kl att (classic)': np.array([-0.347065, -0.530416, -0.576544]),
    'prior Î²-vae, dropout, kl att (quad)': np.array([-0.382611, -0.499135, -0.600870]),
}

# ãƒ—ãƒ­ãƒƒãƒˆè¨­å®š
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# ã‚¹ã‚¿ã‚¤ãƒ«ã¨ãƒžãƒ¼ã‚«ãƒ¼ã®è¨­å®š
styles = {
    'Î²-vae': {'linestyle': '--', 'marker': None, 'color': None},
    'prior Î²-vae, dropout, kl att (classic)': {'linestyle': '-', 'marker': 'o', 'color': None},
    'prior Î²-vae, dropout, kl att (quad)': {'linestyle': '-', 'marker': 'x', 'color': None},
    'zero prior, kl att (quad)': {'linestyle': '-', 'marker': 'x', 'color': None},
}

# å·¦ã®ãƒ—ãƒ­ãƒƒãƒˆï¼šTest Reconstruction Loss
for method, values in test_recon.items():
    style = styles[method]
    ax1.plot(beta_values, values, 
             label=method, 
             linestyle=style['linestyle'],
             marker=style['marker'],
             markersize=8)

ax1.set_xlabel('beta')
ax1.set_ylabel('test recon loss')
ax1.set_xscale('log')
ax1.grid(True, alpha=0.3)
ax1.set_title('test recon loss')

# å³ã®ãƒ—ãƒ­ãƒƒãƒˆï¼šGeneralization Gap
for method, values in gen_gap.items():
    style = styles[method]
    ax2.plot(beta_values, values, 
             label=method, 
             linestyle=style['linestyle'],
             marker=style['marker'],
             markersize=8)

ax2.set_xlabel('beta')
ax2.set_ylabel('gen gap')
ax2.set_xscale('log')
# ax2.set_yscale('log')
ax2.grid(True, alpha=0.3)
ax2.set_title('gen gap')

# å‡¡ä¾‹
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)

plt.tight_layout()
plt.savefig('output_graph.pdf')  # PDFã¨ã—ã¦ä¿å­˜
plt.show()
```
